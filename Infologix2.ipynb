```python
# ==========================================================
# PROJET COPILOTE – ANALYSE DE TRACES UTILISATEURS
# Julien Durand - Étapes : compréhension, chargement, exploration, features
# ==========================================================

import pandas as pd
import warnings
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import math
import csv
import re
from collections import Counter

# -------------------------------------------------------------------
# Étape 1 : Chargement des données brutes en mémoire
# -------------------------------------------------------------------

def read_ds(ds_name: str):
    """
    Lit un fichier CSV contenant des traces d'utilisation de Copilote.
    Chaque ligne contient : utilisateur, navigateur, puis une série d'actions.
    """
    path = fr"C:\\Users\\Julien\\Documents\\Centrale Lyon 20252026\\Data science\\CSV\\{ds_name}.csv"

    df = pd.read_csv(
        path,
        header=None,
        sep=",",
        encoding="latin1",      # pour corriger les caractères accentués
        engine="python",
        on_bad_lines="skip",
        quoting=csv.QUOTE_NONE
    )

    # Renommage des deux premières colonnes
    df.rename(columns={0: 'util', 1: 'navigateur'}, inplace=True)
    return df

# Chargement des fichiers train et test
features_train = read_ds("train")
features_test = read_ds("test")

print("Forme des datasets :", features_train.shape, features_test.shape)
print("\nAperçu des premières lignes :")
print(features_train.head(3))

# -------------------------------------------------------------------
# Étape 2 : Première analyse des données
# -------------------------------------------------------------------

# Aperçu partiel (les 20 premières colonnes)
print("\nAperçu partiel des premières colonnes :")
print(features_train.loc[:, :20].head())

# Vérification des navigateurs distincts
print("\nValeurs distinctes dans la colonne 'navigateur' :")
print(features_train['navigateur'].value_counts(dropna=False))

# -------------------------------------------------------------------
# Fonctions utilitaires (warnings, affichage Markdown, graphiques)
# -------------------------------------------------------------------

def ignore_warnings(f):
    def _f(*args, **kwargs):
        warnings.filterwarnings('ignore')
        v = f(*args, **kwargs)
        warnings.filterwarnings('default')
        return v
    return _f

def markdown_table(headNtail=False, use_index=True, title=None, precision=2):
    """
    Décorateur pour afficher un DataFrame Pandas sous forme de tableau Markdown.
    """
    def _get_value(val): return str(round(val, precision) if isinstance(val, float) else val)
    def _format_row(row): 
        row_str = ""
        if use_index: row_str += f"|{str(row.name)}"
        for value in row.values: row_str += f"| {_get_value(value)}"
        return row_str + "|"
    def _get_str(df):
        return "\n".join(df.apply(_format_row, axis=1))
    def _deco(f):
        def _f(*args, **kwargs):
            df = f(*args, **kwargs)
            _str = f"#### {title}\n" if title else ""
            header = ([str(df.index.name)] if use_index else []) + df.columns.astype(str).to_list() 
            _str += f"|{'|'.join(header)}|" + f"\n|{'--|'*len(header)}\n" if header else None
            if headNtail:
                _str += _get_str(df.head())
                _str += "\n|...|...|\n"
                _str += _get_str(df.tail())
            else:
                _str += _get_str(df)
            display(Markdown(_str))
        return _f
    return _deco

def get_grid(n, n_row=None, n_col=None, titles=None, figsize=(10, 8), wspace=.5, hspace=.5, **kwargs):
    if n_row: n_col= n_col or math.floor(n/n_row)
    elif n_col: n_row= n_row or math.ceil(n/n_col)
    else:
        n_row = math.ceil(math.sqrt(n))
        n_col = math.floor(n/n_row)
    fig, axs = plt.subplots(n_row, n_col, figsize=figsize, **kwargs)
    plt.subplots_adjust(hspace=hspace, wspace=wspace)
    if titles is not None:
        for ax, title in zip(axs.flat, titles): ax.set_title(title)
    return fig, axs

# -------------------------------------------------------------------
# Étape 3 : Statistiques exploratoires simples
# -------------------------------------------------------------------

@markdown_table(title="Navigateur par utilisateur", headNtail=True)
def browsers_per_player(df):
    return pd.crosstab(df['util'], df['navigateur'])

browsers_per_player(features_train)

# -------------------------------------------------------------------
# Étape 4 : Construction de caractéristiques (feature engineering)
# -------------------------------------------------------------------

def filter_action(value: str):
    """Nettoie une action en supprimant les détails après (, <, $, etc."""
    for delim in ["(", "<", "$", "1"]:
        if delim in value and (low_ind := value.index(delim)):
            value = value[:low_ind]
    return value.strip()

# Conversion des actions en une seule liste par utilisateur
actions_only = features_train.iloc[:, 2:].stack().dropna().unique()
filtered_actions = sorted(set([filter_action(a) for a in actions_only if not a.startswith("t")]))
print(f"\nNombre d’actions uniques filtrées : {len(filtered_actions)}")
print(filtered_actions[:15])

# --- Extraction d’informations supplémentaires à partir des regex ---
pattern_ecran = re.compile(r"\((.*?)\)")
pattern_conf_ecran = re.compile(r"<(.*?)>")
pattern_chaine = re.compile(r"\$(.*?)\$")

def extract_features(row):
    """Extrait des informations utiles de la ligne brute."""
    actions = row.dropna().values[2:]
    actions_str = ",".join(map(str, actions))
    
    # Comptage simple
    nb_actions = len(actions)
    nb_t = actions_str.count("t")
    
    # Écrans et chaînes les plus fréquents
    ecrans = pattern_ecran.findall(actions_str)
    chaines = pattern_chaine.findall(actions_str)
    
    ecran_most = Counter(ecrans).most_common(1)[0][0] if ecrans else None
    chaine_most = Counter(chaines).most_common(1)[0][0] if chaines else None
    
    return pd.Series({
        "nb_actions": nb_actions,
        "nb_t": nb_t,
        "ecran_most": ecran_most,
        "chaine_most": chaine_most
    })

features_train_extracted = features_train.apply(extract_features, axis=1)
features_train_clean = pd.concat([features_train[['util', 'navigateur']], features_train_extracted], axis=1)

print("\nExemple de features construites :")
print(features_train_clean.head())

# -------------------------------------------------------------------
# Étape 5 : Affichage Markdown et distributions
# -------------------------------------------------------------------

@markdown_table(title="Dataset complet (train)")
def show_full_dataset(df):
    """Affiche tout le DataFrame au format Markdown (⚠️ lent si très grand)."""
    return df

@markdown_table(headNtail=True, title="Distribution d'une variable")
def variable_distribution(df, col):
    """Affiche la distribution (valeurs uniques et effectifs) d'une variable."""
    if col not in df.columns:
        print(f"La colonne '{col}' n'existe pas dans le DataFrame.")
        return pd.DataFrame()
    return df[col].value_counts().to_frame(name="count")


# show_full_dataset(features_train_clean)
# variable_distribution(features_train_clean, "navigateur")

# -------------------------------------------------------------------
# Étape 6 : Statistiques simples sur les variables construites
# -------------------------------------------------------------------

print("\nStatistiques descriptives :")
print(features_train_clean.describe())

# Distribution du nombre d’actions
plt.figure(figsize=(8,5))
plt.hist(features_train_clean["nb_actions"], bins=50)
plt.title("Distribution du nombre d’actions par session")
plt.xlabel("Nombre d’actions")
plt.ylabel("Fréquence")
plt.show()

# Répartition des navigateurs
plt.figure(figsize=(6,4))
features_train_clean["navigateur"].value_counts().plot(kind='bar')
plt.title("Répartition des navigateurs")
plt.xlabel("Navigateur")
plt.ylabel("Nombre d’utilisateurs")
plt.show()

print("\nExploration terminée ✅ (jusqu’à la construction de features).")

# -------------------------------------------------------------------
# Étape 7 : Traitements préliminaires et nettoyage
# -------------------------------------------------------------------

# Vérification des valeurs aberrantes (outliers)
print("\nVérification des valeurs aberrantes :")
print(features_train_clean.describe())

# Suppression simple des outliers (au-delà de 3 écarts-types)
numeric_cols = ["nb_actions", "nb_t"]
for col in numeric_cols:
    mean = features_train_clean[col].mean()
    std = features_train_clean[col].std()
    mask = (features_train_clean[col] >= mean - 3*std) & (features_train_clean[col] <= mean + 3*std)
    features_train_clean = features_train_clean[mask]

print("\nTaille du dataset après suppression des outliers :", features_train_clean.shape)
# -------------------------------------------------------------------
# Étape 8 : Encodage de la variable dépendante (util)
# -------------------------------------------------------------------

features_train_clean["util"] = pd.Categorical(features_train_clean["util"])

def to_categories(df, col="util"):
    df[col] = df[col].cat.codes
    return df

features_train_clean = to_categories(features_train_clean, "util")
print("\nConversion de la variable 'util' en catégories numériques effectuée ✅")
print(features_train_clean.head())
# -------------------------------------------------------------------
# Étape 9 : Définition de nouvelles features à partir des données brutes
# -------------------------------------------------------------------

def filter_action(value: str):
    for delim in ["(", "<", "$", "1"]:
        if delim in value and (low_ind := value.index(delim)):
            value = value[:low_ind]
    return value.strip()

uniques = features_train.iloc[:, 2:].stack().dropna().unique()
filtered_uniques = list(set([filter_action(u) for u in uniques if not u.startswith("t")]))
print(f"\nNombre d’actions filtrées uniques : {len(filtered_uniques)}")

# Création de features avancées
pattern_ecran = re.compile(r"\((.*?)\)")
pattern_conf_ecran = re.compile(r"<(.*?)>")
pattern_chaine = re.compile(r"\$(.*?)\$")

def enrich_features(row):
    actions = row.dropna().values[2:]
    actions_str = ",".join(map(str, actions))

    nb_actions = len(actions)
    nb_t = actions_str.count("t")
    nb_click = actions_str.count("click")
    nb_scroll = actions_str.count("scroll")
    ratio_t = nb_t / nb_actions if nb_actions else 0
    
    ecrans = pattern_ecran.findall(actions_str)
    chaines = pattern_chaine.findall(actions_str)
    confs = pattern_conf_ecran.findall(actions_str)

    ecran_most = Counter(ecrans).most_common(1)[0][0] if ecrans else None
    chaine_most = Counter(chaines).most_common(1)[0][0] if chaines else None
    conf_most = Counter(confs).most_common(1)[0][0] if confs else None

    return pd.Series({
        "nb_actions": nb_actions,
        "nb_t": nb_t,
        "nb_click": nb_click,
        "nb_scroll": nb_scroll,
        "ratio_t": ratio_t,
        "ecran_most": ecran_most,
        "chaine_most": chaine_most,
        "conf_most": conf_most
    })

features_engineered = features_train.apply(enrich_features, axis=1)
features_final = pd.concat([features_train[['util', 'navigateur']], features_engineered], axis=1)

print("\nAperçu des features enrichies :")
print(features_final.head())
# -------------------------------------------------------------------
# Étape 10 : Encodage des chaînes de caractères
# -------------------------------------------------------------------

# Option 1 : variable catégorielle
features_final["navigateur_cat"] = pd.Categorical(features_final["navigateur"]).codes

# Option 2 : One-Hot Encoding (utile pour les modèles linéaires)
features_final = pd.get_dummies(features_final, columns=["navigateur"], prefix="nav", drop_first=True)

print("\nEncodage des navigateurs effectué ✅")
print(features_final.head())
# -------------------------------------------------------------------
# Étape 11 : Exploration et visualisation des caractéristiques
# -------------------------------------------------------------------

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=False, figsize=(18,6))

ax1.hist(features_final["nb_actions"], bins=30)
ax1.set_title("Distribution du nombre d'actions")
ax1.set_xlabel("nb_actions")

ax2.hist(features_final["nb_t"], bins=30)
ax2.set_title("Distribution du nombre de 't'")
ax2.set_xlabel("nb_t")

ax3.hist(features_final["ratio_t"], bins=30)
ax3.set_title("Ratio t / nb_actions")
ax3.set_xlabel("ratio_t")

plt.show()

# Interprétation :
# - nb_actions mesure la longueur moyenne d'une session utilisateur
# - nb_t correspond au nombre d'événements 't' (peut indiquer le temps passé)
# - ratio_t donne une idée du profil d'utilisation
# -------------------------------------------------------------------
# Étape 12 : Corrélation entre caractéristiques et classes
# -------------------------------------------------------------------

def plot_categories(categories, type_="box", feature="nb_actions"):
    fig, axs = get_grid(len(categories), n_row=1, figsize=(16, 4))
    for cat, ax in zip(categories, axs.flat):
        subset = features_final[features_final["util"] == cat]
        if type_ == "box":
            ax.boxplot(subset[feature])
        elif type_ == "violin":
            ax.violinplot(subset[feature])
        ax.set_title(f"Catégorie {cat}")
        ax.set_xlabel(feature)
    plt.show()

categories = [0, 15, 152, 199]
plot_categories(categories, type_="box", feature="nb_actions")
```
