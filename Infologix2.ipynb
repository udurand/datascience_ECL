```python
# ==========================================================
# PROJET COPILOTE – ANALYSE DE TRACES UTILISATEURS
# Julien Durand – Centrale Lyon
# Étapes : compréhension, chargement, exploration, features, classification
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import warnings
import math
import csv
import re
from collections import Counter
from IPython.display import display, Markdown

# ==========================================================
# Étape 1 : Chargement des données brutes
# ==========================================================

def read_ds(ds_name: str):
    """
    Lit un fichier CSV contenant des traces utilisateurs.
    Chaque ligne : utilisateur, navigateur, puis une série d'actions.
    """
    path = fr"C:\\Users\\Julien\\Documents\\Centrale Lyon 20252026\\Data science\\CSV\\{ds_name}.csv"

    with open(path, encoding="latin1") as f:
        max_actions = max((len(str(c).split(",")) for c in f.readlines()))
        f.seek(0)
        _names = ["util", "navigateur"] if "train" in ds_name else ["navigateur"]
        _names.extend(range(max_actions - len(_names)))
        return pd.read_csv(f, names=_names, dtype=str, on_bad_lines='skip')

# Chargement des fichiers train et test
features_train = read_ds("train")
features_test = read_ds("test")

print("✅ Données chargées :")
print("Forme train / test :", features_train.shape, features_test.shape)
print("\nAperçu des premières lignes :")
display(features_train.head())

# ==========================================================
# Étape 2 : Première analyse des données
# ==========================================================

print("\nAperçu partiel des 20 premières colonnes :")
display(features_train.loc[:, :20].head())

print("\nNombre total de colonnes :", features_train.shape[1])
print("Valeurs distinctes de la colonne 'navigateur' :")
print(features_train["navigateur"].value_counts(dropna=False))

# ==========================================================
# Étape 3 : Fonctions utilitaires
# ==========================================================

def ignore_warnings(f):
    """Décorateur pour ignorer temporairement les warnings."""
    def _f(*args, **kwargs):
        warnings.filterwarnings('ignore')
        v = f(*args, **kwargs)
        warnings.filterwarnings('default')
        return v
    return _f

def markdown_table(headNtail=False, use_index=True, title=None, precision=2):
    """Affiche un DataFrame sous forme de tableau Markdown."""
    def _get_value(val): return str(round(val, precision) if isinstance(val, float) else val)
    def _format_row(row):
        row_str = ""
        if use_index: row_str += f"|{str(row.name)}"
        for value in row.values: row_str += f"| {_get_value(value)}"
        return row_str + "|"
    def _get_str(df):
        return "\n".join(df.apply(_format_row, axis=1))
    def _deco(f):
        def _f(*args, **kwargs):
            df = f(*args, **kwargs)
            _str = f"#### {title}\n" if title else ""
            header = ([str(df.index.name)] if use_index else []) + df.columns.astype(str).to_list()
            _str += f"|{'|'.join(header)}|" + f"\n|{'--|'*len(header)}\n"
            if headNtail:
                _str += _get_str(df.head())
                _str += "\n|...|...|\n"
                _str += _get_str(df.tail())
            else:
                _str += _get_str(df)
            display(Markdown(_str))
        return _f
    return _deco

def get_grid(n, n_row=None, n_col=None, titles=None, figsize=(10, 8), wspace=.5, hspace=.5, **kwargs):
    """Crée une grille de sous-graphiques."""
    if n_row: n_col = n_col or math.floor(n/n_row)
    elif n_col: n_row = n_row or math.ceil(n/n_col)
    else:
        n_row = math.ceil(math.sqrt(n))
        n_col = math.floor(n/n_row)
    fig, axs = plt.subplots(n_row, n_col, figsize=figsize, **kwargs)
    plt.subplots_adjust(hspace=hspace, wspace=wspace)
    if titles is not None:
        for ax, title in zip(axs.flat, titles): ax.set_title(title)
    return fig, axs

# ==========================================================
# Étape 4 : Exploration – statistiques descriptives
# ==========================================================

@markdown_table(title="Navigateur par utilisateur", headNtail=True)
def browsers_per_player(df):
    return pd.crosstab(df["util"], df["navigateur"])

browsers_per_player(features_train)

@markdown_table(headNtail=True, title="Stats: Y distribution")
def get_Y_stats(df):
    return df["util"].value_counts().to_frame(name="count")

get_Y_stats(features_train)

# ==========================================================
# Étape 5 : Pré-traitements et construction des features
# ==========================================================

# --- Suppression d’éventuels outliers ---
# TODO: inspecter s'il y a des valeurs aberrantes et les retirer si nécessaire

# --- Conversion de la variable cible en catégories numériques ---
features_train["util"] = pd.Categorical(features_train["util"])

def to_categories(df, col="util"):
    df[[col]] = df[[col]].apply(lambda x: x.cat.codes)

to_categories(features_train)

# --- Extraction des actions uniques ---
def filter_action(value: str):
    """Nettoie une action en supprimant les détails inutiles."""
    for delim in ["(", "<", "$", "1"]:
        if delim in value and (low_ind := value.index(delim)):
            value = value[:low_ind]
    return value.strip()

uniques = features_train.iloc[:, 2:].stack().unique()
filtered_uniques = list(set([filter_action(un) for un in uniques if not un.startswith("t")]))
print(f"\nNombre d’actions uniques filtrées : {len(filtered_uniques)}")
print(filtered_uniques[:15])

# ==========================================================
# Étape 6 : Extraction de nouvelles caractéristiques
# ==========================================================

pattern_ecran = re.compile(r"\((.*?)\)")
pattern_conf_ecran = re.compile(r"<(.*?)>")
pattern_chaine = re.compile(r"\$(.*?)\$")

def extract_features(row):
    """Extrait des informations à partir d'une ligne brute."""
    actions = row.dropna().values[2:]
    actions_str = ",".join(map(str, actions))
    nb_actions = len(actions)
    nb_t = actions_str.count("t")
    ecrans = pattern_ecran.findall(actions_str)
    chaines = pattern_chaine.findall(actions_str)
    ecran_most = Counter(ecrans).most_common(1)[0][0] if ecrans else None
    chaine_most = Counter(chaines).most_common(1)[0][0] if chaines else None
    return pd.Series({
        "nb_actions": nb_actions,
        "nb_t": nb_t,
        "ecran_most": ecran_most,
        "chaine_most": chaine_most
    })

features_train_extracted = features_train.apply(extract_features, axis=1)
features_train_clean = pd.concat([features_train[["util", "navigateur"]], features_train_extracted], axis=1)

# ==========================================================
# Étape 7 : Traitement des variables catégorielles
# ==========================================================

# Option 1 : variable catégorielle
features_train_clean["navigateur"] = pd.Categorical(features_train_clean["navigateur"])

# Option 2 (alternative) : One-Hot Encoding
# features_train_clean = pd.get_dummies(features_train_clean, columns=["navigateur"])

# ==========================================================
# Étape 8 : Statistiques simples et visualisation
# ==========================================================

print("\nStatistiques descriptives :")
display(features_train_clean.describe())

# Histogrammes de features
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=True, figsize=(20,6))
ax1.hist(features_train_clean["nb_actions"], bins=40)
ax1.set_title("Distribution du nombre d’actions")
ax2.hist(features_train_clean["nb_t"], bins=40)
ax2.set_title("Distribution des marqueurs temporels")
features_train_clean["navigateur"].value_counts().plot(kind="bar", ax=ax3)
ax3.set_title("Répartition des navigateurs")
plt.show()

# ==========================================================
# Étape 9 : Corrélation et visualisation par classes
# ==========================================================

def plot_categories(categories, type_="box", feature="nb_actions"):
    fig, axs = get_grid(len(categories), n_row=1, figsize=(16, 4))
    for i, cat in enumerate(categories):
        subset = features_train_clean[features_train_clean["util"] == cat]
        if type_ == "box":
            axs.flat[i].boxplot(subset[feature].dropna())
        elif type_ == "violin":
            axs.flat[i].violinplot(subset[feature].dropna())
        axs.flat[i].set_title(f"Catégorie {cat}")
    plt.show()

categories = [0, 15, 152, 199]
plot_categories(categories)

print("\nExploration et feature engineering terminés ✅ – prêt pour la modélisation.")
```
